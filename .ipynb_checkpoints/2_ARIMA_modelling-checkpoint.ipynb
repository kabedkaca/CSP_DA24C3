# ARIMA Modelling 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data/household_power_consumption_cleaned_datetime_index_drop_intensity.csv')
df.head()
```

```python
# Set the datetime column as the index
df.set_index('DateTime', inplace=True)
df.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Filter data up to '2007-08'
df_filtered = df[:'2007-08']

# Group by the datetime index and calculate the mean of 'Global_active_power'
df_grouped = df_filtered['Global_active_power'].resample('D').mean()  # Resample by day and calculate the mean

# Plot the autocorrelation function
plot_acf(df_grouped, lags=100, title="Global Active Power Autocorrelation")
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.grid()
plt.show()
```

```python
## PLOTTING AUTO CORRELATION AND PARTIAL CORRELATION
plot_acf(df.groupby(df['DateTime']).mean()['Global_active_power'][:'2007-08'],lags=100,title="Global Reactive Power")
plt.xlabel('lags')
plt.title('Autocorrelation')
plt.grid()
plt.show()
```

```python
# Ensure the index is of datetime type
df.index = pd.to_datetime(df.index)

# Filter data up to '2007-08'
df_filtered = df.loc[:'2007-08']

# Group by the datetime index and calculate the mean of 'Global_active_power'
df_grouped = df_filtered['Global_active_power'].resample('D').mean()

# Plot the autocorrelation function
plot_acf(df_grouped.dropna(), lags=100, title="Global Active Power Autocorrelation")
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.grid()
plt.show()

#KeyError: 'Value based partial slicing on non-monotonic DatetimeIndexes with non-existing keys is not allowed.'
#indicates that the datetime index is not sorted. To resolve this, you need to sort the index before filtering.
```

```python
# Ensure the index is of datetime type
df.index = pd.to_datetime(df.index)

# Sort the index
df = df.sort_index()

# Filter data up to '2007-08'
df_filtered = df.loc[:'2007-08']

# Group by the datetime index and calculate the mean of 'Global_active_power'
df_grouped = df_filtered['Global_active_power'].resample('D').mean()

# Plot the autocorrelation function
plot_acf(df_grouped.dropna(), lags=100, title="Global Active Power Autocorrelation")
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.grid()
plt.show()
```

```python
# Slice the DataFrame up to '2007-08'
act_power_df = df['Global_active_power'][:'2007-08']

# Fit ARIMA model
model1 = ARIMA(act_power_df, order=(5, 2, 0))
model1_fit = model1.fit()

# Make predictions
preds = model1_fit.predict(start=0, end=len(act_power_df) - 1, typ='levels')

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(preds, linewidth=3)
act_power_df.rolling(5).mean().plot(linewidth=2)
plt.legend(['Predicted', 'Observed'])
plt.xlabel('Date')
plt.title('Global Active Power')
plt.show()

#ValueWarning: A date index has been provided, but it has no associated frequency information and so will be ignored when e.g. forecasting.
#  self._init_dates(dates, freq)
#MemoryError: Unable to allocate 84.5 MiB for an array with shape (7, 7, 226081) and data type float64
```
