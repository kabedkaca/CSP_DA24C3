## Data Munging

### change `DATE_BEGIN` and `DATE_END` to datetime

**for code below to change date_begin to datetime format,**

ValueError: unconverted data remains when parsing with format "%Y%m%d%H%M": ".0", at position 66249. You might want to try:
    - passing `format` if your strings have a consistent format;
    - passing `format='ISO8601'` if your strings are all ISO8601 but not necessarily in exactly the same format;
    - passing `format='mixed'`, and the format will be inferred for each element individually. You might want to use `dayfirst` alongside this.
    
reason:
- The error you encountered suggests that there are some inconsistencies in the date format of your DATE_BEGIN column. Specifically, it looks like there are some extra characters (e.g., ".0") at the end of some of the date strings, which is causing the pd.to_datetime() function to fail.
- To handle this, you can try the following approaches:
>Remove any extraneous characters from the date strings before converting to datetime.
?Use the errors='coerce' parameter to convert invalid parsing to NaT.

```python
#Step 1: Clean the Date Strings
# Remove any extraneous characters from the date strings
# df2['DATE_BEGIN'] = df2['DATE_BEGIN'].astype(str).str.replace(r'\.0$', '', regex=True)  
# Remove any extraneous characters from the date strings in both DATE_BEGIN and DATE_END columns
df2[['DATE_BEGIN', 'DATE_END']] = df2[['DATE_BEGIN', 'DATE_END']].apply(lambda col: col.astype(str).str.replace(r'\.0$', '', regex=True))
```

```python
df2['DATE_BEGIN'] = pd.to_datetime(df2['DATE_BEGIN'], errors='coerce')
df2['DATE_END'] = pd.to_datetime(df2['DATE_END'], errors='coerce')
#By using errors='coerce', any date strings that still can't be parsed after cleaning will be converted to NaT (Not a Time), 
#which you can then handle appropriately (e.g., by filling with a default value or dropping those rows).
df2.head()
```

```python
null_begin = df2['DATE_BEGIN'].isnull().sum()
null_end = df2['DATE_END'].isnull().sum()
print('there are', null_begin, 'rows of null in DATE_BEGIN and', null_end, 'rows of null in DATE_END')
```

## Analyze the null vs rows for each column

```python
def analyze_dataframe(df):
    # new dataframe to store new df just for our ref
    result_df2 = pd.DataFrame(index=df.columns, columns=['Unique Values', 'Missing Values', 'Total Rows'])
    # Get the total number of rows
    total_rows = df.shape[0]
    # now the new df must be filled with info for the column
    for column in df.columns:
        unique_values = df[column].nunique()  #to call unique value from each column
        missing_values = df[column].isnull().sum()  #to check how many null for each row 

        result_df2.loc[column] = [unique_values, missing_values, total_rows] #to loc whole colun and rows into result_df2
    return result_df2
```

```python
analyze_dataframe(df2)
```

### HOW TO DEAL WITH NULL IN OTHER COLUMNS

1. Imputation:
If columns have numerical data, impute missing values with the mean, median, or a specific value that makes sense for analysis.
2. Categorical Imputation:
For categorical columns like 'STATE',impute missing values with the mode (most frequent value).
3. Drop Columns with High Null Percentage:
If a column has a very high percentage of null values and it's not critical for analysis, dropping the column altogether.
4. Flagging Null Values:
Instead of removing null values, create a new boolean column indicating whether the value was missing in the original column

For lat lon,that share the same state, i would choose the most frequent lat lon for that state, to fill the null for lat lon,
first let fill the null in `STATE` with the mode, before proceed with the `lat` and `LON`

```python
# Function to fill null values in the STATE column with the mode
def fill_state_with_mode(df2, column='STATE'): #dataframe kita, satet column kita nakrun def ni
    # cari dulu mode untuk column tu
    mode_value = df2[column].mode()[0]

    # Fill null values with the mode
    df2[column].fillna(mode_value, inplace=True)
    
    #return df2

fill_state_with_mode(df2, column='STATE')
```

```python
# def to fill missing lat/lon with the most frequent to that state
def fill_missing_lat_lon(df):
    # Define a function to fill missing values with the most frequent non-null value
    def fill_with_most_frequent(series): #series as we only want to make change on row of either lat or lon
        most_frequent = series.mode()  #to choose the mode for that series
        
        if len(most_frequent) > 0:  #pastikan series.mode() tu ada nilai dia, takut tak wujud lol
            return series.fillna(most_frequent[0])  #fillna: used to fill missing values in series
        else:
            return series  #the series will keep in null

    # Apply the function to fill lat and lon for each state
    df2['LON'] = df2.groupby('STATE')['LON'].transform(fill_with_most_frequent)  #utk row lon, grup ikut state,  lepas tu set utk fill null dgn lon state tu yg paling kerap
    df2['LAT'] = df2.groupby('STATE')['LAT'].transform(fill_with_most_frequent)
    
    #return df2
```
